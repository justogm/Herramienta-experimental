```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("datos/movimientos.csv")
df["t"] = range(1, len(df)+1)
df.info()
```

```python
fig = px.line(df, x="x", y="y", animation_frame="t", range_x=[0, 1920], range_y=[0,1080])
```

```python
df["t"] = str(df["t"])
```

```python
fig.write_html("animacion.html")
```
